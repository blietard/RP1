```python
from sklearn.svm import SVC
from sklearn.svm import LinearSVC

def AccuracyRateBan(classifier,Xtrain,Ttrain,Xbanned,Tbanned,pca=True):
    if pca:
        model = make_pipeline(Scaler(),PCA(n_components=0.975),classifier)
    else:
        model = classifier
    model.fit(Xtrain,Ttrain)
    y_train = model.predict(Xtrain)
    Tacc = np.sum(Ttrain == y_train)/len(Xtrain)
    y_banned = model.predict(Xbanned)
    Bacc = np.sum(Tbanned == y_banned)/len(Xbanned)
    return (Tacc,Bacc)

def PCAAvgAccBan(classifier,X,targets,nouns,pca=True):
    Taccs = np.zeros(len(listNouns))
    Baccs = np.zeros(len(listNouns))
    for i,bannedW in enumerate(listNouns):
        Xtrain = X[nouns!=bannedW]
        Ttrain = targets[nouns!=bannedW]
        Xbanned = X[nouns==bannedW]
        Tbanned = targets[nouns==bannedW]
        Tacc,Bacc = AccuracyRateBan(classifier,Xtrain,Ttrain,Xbanned,Tbanned,pca)
        Taccs[i],Baccs[i] = Tacc, Bacc
    AvgTrain = np.mean(Taccs)
    AvgBanned = np.mean(Baccs)
    return (AvgTrain,AvgBanned)



```

```python
#WITH RAW DATA
print('----------------------------------RAW DATA-----------------------')
classifier = LinearSVC(max_iter=100000)
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=False)
print('SVM(linear) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='poly',degree=3)
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=False)
print('SVM(d=3) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='poly',degree=5)
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=False)
print('SVM(d=5) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='rbf')
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=False)
print('SVM(rbf) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

# With RAW DATA + PCA
print('----------------------------RAW DATA + PCA ---------------------')
classifier = LinearSVC(max_iter=100000)
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=True)
print('SVM(linear) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='poly',degree=3)
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=True)
print('SVM(d=3) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='poly',degree=5)
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=True)
print('SVM(d=5) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='rbf')
Tacc,Bacc =  PCAAvgAccBan(classifier,X3,targets3,nouns3,pca=True)
print('SVM(rbf) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

#With TSNE Data
print('----------------------------------TSNE DATA-----------------------')
df_tsne = pd.DataFrame(data=pcaX3.T,index=X3.index)

classifier = LinearSVC(max_iter=100000)
Tacc,Bacc =  PCAAvgAccBan(classifier,df_tsne,targets3,nouns3,pca=False)
print('SVM(linear) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))


classifier = SVC(kernel='poly',degree=3)
Tacc,Bacc =  PCAAvgAccBan(classifier,df_tsne,targets3,nouns3,pca=False)
print('SVM(d=3) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='poly',degree=5)
Tacc,Bacc =  PCAAvgAccBan(classifier,df_tsne,targets3,nouns3,pca=False)
print('SVM(d=5) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))

classifier = SVC(kernel='rbf')
Tacc,Bacc =  PCAAvgAccBan(classifier,df_tsne,targets3,nouns3,pca=False)
print('SVM(rbf) ------ Train Accuracy : {}  ||  Test Accuracy : {}'.format(np.round(Tacc,4),np.round(Bacc,4)))
```
